### 4. 데이터 프레임 조작
 - 데이터를 가공하는 과정을 데이터 전처리나 데이터 조작이라고 한다.
 - 빅데이터 분석의 단계 중 가장 많은 시간이 소요되는 과정
 - 데이터 사이언티스트 업무 시간 중 80% 정도를 데이터 수집 및 전처리 과정에 사용
 - 데이터 분석 업무 중 가장 싫으 과정

#### 4.1 데이터프레임의 행과 열
 - 변수(Variable, 열, 컬럼, 피처, 특징): 키, 몸무게, 성별
 - 관측지(Observation, 행, 로우,데이터): 값을 측정한 단위, 각각의 사람
 - 값(Value): 152cm, 80kg, 여성
 
#### 4.2 행과 열의 크기
 - 데이터프레임에서 행이 많다는 의미
 - 행의 증가는 막대한 컴퓨팅 파워의 증가를 의미
  1. 행이 늘어나면 컴퓨터의 연산 핵심인 메모리와 cpu를 늘려야함
  2. 그게 어려우면 클라우드 컴퓨팅의 파워를 빌려야함(분석 기법은 큰 차이없음)
 - 열의 감소는 분석시 다양한 조합이 불가능(처음시작시 열을 많이 늘여놓으면 좋다)

#### 4.3 깔끔한 데이터
 - 데이터를 조작하고, 모향화하고, 시각화가 용이함
 - 특정한 구조를 갖추고 있는데 변수는 열, 관측점은 행이며, 관측 단위에 대한 형태는 테이블로 구성
 - 데이터 분석을 쉽게 할 수 있는 데이터

#### 4.4-6 데이터 합치기

```python
# 시리즈 데이터 합하기
# 시리즈 변수명 s1 ==> 값은 0,1 /인덱스 값은 A,B로 시리즈 만들기
# 시리즈 변수명 s2 ==> 값은 2,3,4 /인덱스 값은 a,b,c로 시리즈 만들기

s1 = pd.Series({'A':0,'B':1})
s2 = pd.Series({'a':2,'b':3,'c':4})
```

```python
# 행단위로 합치기(열의갯수가 같아야함)
pd.concat([s1,s2], axis = 0)
```
