## 빅데이터 분석과 머신러닝

### 데이터 탐색
 - 데이터 전처리 요소 중 하나, 데이터 분석을 위한 다양한 관점을 제공
 - 데이터 그룹 분석
 - 데이터 재구조화
 - 시계열 데이터

### 그룹분석 
 - 특정한 조건에 맞는 데이터가 하나이상 데이터 그룹을 이루는 경우 그룹의 특성을 보여주는 그룹분석을 자주 사용
 - (주로 범주형 데이터를 묶은후 사용하게 됨)
 - 그룹분석은 분석 대상 변수를 그룹별로 데이터를 집계하여 진행하는 분석, 데이터분석에서 자주 사용하기 때문에 반드시 알아야할 필수적 방법
 - 기준(범주형)에 따라 여러그룹으로 데이터를 분할(splitting) -반별 그룹으로 분할
 - 각 그룹에 독립적으로 함수를 적용(applying) - 반별로 평균 산정(적용)
 - 결과물들을 하나의 데이터 구조로 결합(combing) - 반별 평균을 결합
 
#### 그룹분석 공식
 - 그룹분석은 대부분 범주형 변수가 그룹연산 기준이 된다.
 - 많은 분석이 범주형과 연속이 결합된 그룹분석으로 진행합니다.

### 데이터 재구조화
#### 데이터 구조화
 - 분석 과정에서 원본 데이터 구조가 분석 기법에 맞지 않아 행과 열의 위치를 바꾼다거나, 특정요인에 따라 집계 해서 구조를 바꿔야하는 경우가 발생
 - 재구조화(Reshaping)라고 하며 판다스는 아래와 같이 재그룹화 기능을 제공함
  1. pd.cut(), pd.qcut(): 데이터 구간화
  2. pd.get_dummies: 원 - 핫 인코딩
   - 원-핫 인코딩: 
  3. T: 데이터 전치
  4. pivot(), pd.pivot_table(): 피봇테이블
  5. stack(), unstack()
  
#### 데이터 구간화
 - 연속형 변수를 바로 사용하는 경우보다 일정 구간(bin이라도도 한다)으로 작업하는 경우
 - 연속형 변수를 버주형 변수로 만드는 방법을 데이터 구간화라고 함
 - 데이터 구간화에도 특별한 원칙이 있는 것이 아니기 때문에 데이터사이언티스트의 업무 이해도에 따라 창의적인 방법으로 할 수 있음
  1. 동일길이로 나누기: pd.cut()
  2. 동일 개수로 나누기: pd.qcut()

#### 원-핫 인코딩
 - 딥러닝 알고리즘은 수치 데이터만 이해할 수 있음
 - 기계가 이해할 수 있는 형태로 데이터를 변환해야함, 이때 범주형 데이터를 원-핫인코딩 형태로 변환함(원-핫 인코딩 -> 범주형 데이터를 0 또는 1의 값으로 변환하는 방법)
 - 원-핫 인코딩
 - 조건에 맞는 하나의 데이터만 1로 변경, 나머지는 0(dummy)으로 변경하는 과정, 아래 예를 보면 동물이라는 컬럼에 사자, 곰, 여우가 있다고 하면, 이때 각각의 동물인 사자, 곰, 여우로 컬럼을 만들고 해당 되는 동물에만 1로 표기 하고 나머지 동물은 0으로 변경
 
#### 데이터 전치(T)
 - 데이터프레임 행과 열의 기준(축)을 바꾸는 방법을 데이터 전치(Transpose)라고 함
 - 수학 선형대수에서 '전치행렬(Transpose matrix)이라는 특수한 형태의 행렬이 있는데 행렬의 내적(inner product)을 구할때, aT*a 처럼 전치행렬(aT)과 원래 행렬(a)을 곱할때 전치 행렬(aT)를 사용함.(행렬은 데이터프레임을 의미함)
 
#### 피봇테이블
 - 피봇테이블이란 많은 양의 데이터에서 필요한 자료만 뽑아 새롭게 데이터를 재구성하는 기능
 - 피봇 테이블을 사용하면 사용자가 원하는 대로 데이터를 정렬하고 필터링 할 수 있음
 - 판다스는 피봇테이블을 만들기 위한 pivot()함수를 제공하는데, 데이터 열 중에서 두 개의 열을 각각 행 인덱스, 열 인덱스로 사용하여 데이터를 조회하여 펼쳐놓은 것을 의미
 - 첫번째 인수로는 행 인덱스로 사용할 열 이름, 두번째 인수로는 열 인덱스로 사용할 열 이름, 그리고 마지막으로 데이터로 사용할 열 이름을 지정
 
#### 멜트
 - 멜트라는 의미처럼 녹으면 흘러내린다
 - 멜트(melt)는 열을 행으로 재구조화 하는 과정
 - 데이터프레임에서는 열이 행으로 흘러 열이 짧아지고 행이 길어진다고 이해하면 됨

#### stack / unstack
 - 행 인덱스와 열 인덱스 교환시 사용하는 기능
 - 스택 명령을 사용하면 열을 행으로 변환하는데 열 인덱스가 반시계 방향으로 90도 회전한 것과 같은 모양이 됨
 - 언스택은 스택으로 쌓은 것을 옆으로 늘여 놓은 것이라 생각하면됨. 둘 다 인덱스를 지정할 때는 문자열 이름과 순서를 표시하는 숫자 인덱스를 모두 사용할 수 있음

### 재구조화

```python
# cut
df1 = pd.cut(df["math"],3)

# 수학점수를 세구간을 만들며
# 동일한 값의 범위(동일한 길이)로 만듬
df1
```

```python
df2 = df['math'].groupby(df1)
```

```python
df2.head(1)
```

```python
# df2에 대한 기초통계(구간별 통계)
df3 = df2.agg(["count","mean","std","min","max"])
```

```python
df3
```

```python
# qcut()
# 수학점수의 갯수를 기준으로 제시된 구간의 값으로
# 유사한 값들끼리(정확하게 같진 않음) 3구간으로 나눔
# 구간이 커지면, 해당 구간이 없는 값이 생길 수 도 있음
# 해당 구간에 존재하는 값이 없을 경우 해당구간은 조회되지 않음
df4 = pd.qcut(df['math'], 3)
```

```python
df4 = pd.qcut(df['math'], 3, labels = False)
# labels = False : 범주의 번호 표시하기
```

```python
df4
```

```python
import numpy as np
df5 = pd.qcut(df['math'],3 , labels = np.arange(3,0,-1))

# 3부터 하나씩, 작아지게 만듬
```
