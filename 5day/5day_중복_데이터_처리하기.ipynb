## 중복데이터 제거
 - 데이터 분석전에 중복 데이터를 확인하고 처리하는 데이터 클린징 작업이 필요
 - 판다스에서 중복 데이터를 확인할 때 duplicated() 함수를 사용
 - 중복값을 처리 할때 drop_duplicates()를 사용

### 중복데이터 처리방법
 - 중복데이터 확인으로 중복이 있으면 무엇을 남길지 확인
 - 중복값 처리(유일한 1개 키만 남기고 나머지 중복제거)
  - DataFrame.drop_duplicates()
 - 중복데이터 결과

### 중복데이터 시작과 끝 확인
 - 중복이 있는 행들 중에서 처음이나 끝에 있는 중복행 중 무엇을 남길지 확인
 - drop_duplicates
 - keep = 'first','last',False 속성에 따라 유일한 1개의 key값만 남기고 나머지 중복제거(삭제)
 - 중복데이터는 행단위로 이루어진다.

### 중복데이터 처리하기

#### 데이터 가져오기

```python
# bicycle_dup.csv읽어 들인후 조회하기
import pandas as pd
```

```python
file_path = "./data/bicycle_dup.csv"
df = pd.read_csv(file_path)
```

```python
df # 420 rows × 12 columns
```

```python
# 중복데이터 확인하기

# 한개의 컬럼을 기준으로 중복체크
# duplicated에 keep을 지정안하면
# 기본값은 중복값 중에 첫번째 값은 False
# 그 외의 중복된 값은 모두 True
# 제거할 때는 True인 행만 제거한다

df.duplicated(["이용거리"])

# True 값이 있다 -> 중복이 있다
```

```python
df['이용거리'].duplicated()
```

```python
# 두개 컬럼을 기준으로 중복체크하기
df[['이용거리','나이']].duplicated()

# 또는 df.duplicated(['이용거리','나이'])
```

```python
# 중복데이터의 조회
df[df[['이용거리','나이']].duplicated()]

# 또는 df.loc[df.duplicated(['이용거리','나이'])]
```

```python
# 중복데이터 - keep = "first"
# keep 생략하면 기본 : keep = first 있다
# 가장 많이 사용하는 중복체크 방식

df.duplicated(["이용거리"], keep = 'first')
```

```python
# keep = "last"
df.duplicated(["이용거리"], keep = 'last')
```

```python
# keep = False
df.duplicated(["이용거리"], keep = False)

# False로 나오는 값은 중복이 없다는 의미
# 중복되는게 하나라도 있는 데이터는 True
```

```python
# 이용거리, 나이 두개 컬럼을 기준으로 중복되는 모든 데이터를 조회
print(len(df[df.duplicated(["이용거리","나이"], keep = False)]))
df_dup = df[df.duplicated(["이용거리","나이"], keep = False)]
```

```python
len(df_dup)
```

### 중복데이터 제거하기

```python
# 첫번째 중복값만 남기고 나머지 중복 데이터 제거하기
# 실행결과는 제거된 후의 정상 데이터만 조회된다.

df.drop_duplicates(["이용거리"], keep = 'first')
```

```python
# 마지막 중복값들만 남기고 나머지 제거하기
df.drop_duplicates(["이용거리"], keep = 'last')
```

```python
# 중복된 데이터 전체 삭제하기
df_dup = df.drop_duplicates(["이용거리"], keep = False)
```

```python
# 중복검증
df_dup[df_dup.duplicated(["이용거리","나이"])]
```

```python
df_dup.describe()
```

```python
# 중복데이터 확인하는 방법
df2 = df.duplicated(["이용거리"], keep = 'first')
```
