```python
# 파일불러오기
df_trf = pd.read_csv("./01_data_원본/포항시_버스_사용자_교통카드_사용_내역/trfcard (0)/trfcard.csv")

# 메타정의서 가져오기
trf_col = pd.read_excel("./01_data_원본/포항시_버스_사용자_교통카드_사용_내역/trfcard (0)/trfcard_columns.xlsx",
                              sheet_name = "컬럼 정의서",
                              usecols = "B,C")

# 컬럼명 데이터를 딕셔너리로 변환
trf_list = trf_col.to_dict("records")

new_dic = {}

for dict_kv in trf_list:
    new_dic[dict_kv["Unnamed: 1"]] = dict_kv["Unnamed: 2"]

new_dic

# 컬럼명 변환
df_trf_kor = df_trf.rename(columns = new_dic)
```

```python
lists = []

for i in range(100) :
    file_path = './01_data_원본/포항시_버스_사용자_교통카드_사용_내역/'
    
    df_cols = pd.read_excel(file_path+'trfcard ({})'.format(i)+'/trfcard_columns.xlsx', 
                            header = 2, 
                            usecols = 'B, C')
    
    df = pd.read_csv(file_path+'trfcard ({})'.format(i)+'/trfcard.csv')

    df_cols_list = df_cols.to_dict('records')

    new_dict = {}

    for dict_0 in df_cols_list :
        new_dict[dict_0['컬럼명 (영문)']] = dict_0['컬럼명 (한글)']

    df = df.rename(columns = new_dict)
    
    lists.append(df)
    
df_m = pd.concat(lists, ignore_index = True)
df_m
```
